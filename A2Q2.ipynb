```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cold(x):
    if x < -10:
        return 0
    elif -10 <= x <= 0:
        return (x + 10) / 10
    elif 0 < x < 5:
        return 1
    else:
        return 0
```

```python
def worm(x):
    if  x < 25:
        return 0
    elif 25 < x < 35:
        return (x-25)/(35-25)
    elif 35< x <40:
        return 1
    else:
        return 0
```

```python
# Calculating membership values for A and B
y_values = np.arange(-10 ,15, 1)
x_values = np.arange(25, 50, 1)
Cold_values = np.array([cold(x) for x in y_values])
Worm_values = np.array([worm(x) for x in x_values])

print("Cold Values:", Cold_values)
print("Warm Values:", Worm_values)


# Performing the intersection operation
intersection_values = np.minimum(Cold_values, Worm_values)


# Find the maximum value from the intersection values
max_intersection_value = np.max(intersection_values)

print("The max-min composition is ", max_intersection_value)
```

```python
# Plotting the results
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(x_values, Worm_values, label='μ_W(x)')
plt.plot(y_values, Cold_values, label='μ_C(x)')
plt.title('Membership Functions')
plt.xlabel('Speed (x)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
```

```python
def max_product_composition(cold_values, warm_values):
    composition_values = []
    for x in x_values:
        max_product = 0
        for y in y_values:
            product = cold(y) * worm(x)
            if product > max_product:
                max_product = product
        composition_values.append(max_product)
    return np.array(composition_values)

composition_values = max_product_composition(Cold_values, Worm_values)
print(composition_values)
max_composition_value = np.max(composition_values)
print("The max-product composition is ", max_composition_value)
```
